```python
import requests

# Define la URL y los par√°metros
url = 'https://serpapi.com/search'
params = {
    'engine': 'google_jobs',
    'q': 'cientifico de datos',  # Reemplaza esto con el t√≠tulo del trabajo que est√°s buscando
    'location': 'Spain',  # Reemplaza esto con la ubicaci√≥n que est√°s buscando
    'api_key': '6212cfc0a2334dc7ff7fe69887340584a1bb21f2d7f5f36d9ef8cd533e222ccf'  # Reemplaza esto con tu clave de API de SerpApi
}

# Haz una solicitud GET a la API
response = requests.get(url, params=params)

# Imprime la respuesta en formato JSON
print(response.json())
```

```python
import pandas as pd

# Extrae la informaci√≥n relevante de los resultados de la b√∫squeda
jobs_data = []
for job in response.json()['jobs_results']:
    job_info = {
        'title': job.get('title'),
        'company_name': job.get('company_name'),
        'location': job.get('location'),
        'description': job.get('description'),
        # Aseg√∫rate de que el campo 'salaries' exista antes de intentar acceder a √©l
        'salaries': job.get('salaries') if 'salaries' in job else None,
    }
    jobs_data.append(job_info)

# Crea un DataFrame con los datos
df = pd.DataFrame(jobs_data)

# Guarda el DataFrame como un archivo CSV
df.to_csv('jobs_data.csv', index=False)
```
